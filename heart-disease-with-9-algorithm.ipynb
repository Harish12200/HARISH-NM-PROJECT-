# x and y split



```python
x= df.drop(['HeartDisease'],axis=1)
y=df['HeartDisease']
```

# spliting x and y into train & test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42,stratify=y)
```

# Model

## Logistics = 73%



```python
from sklearn.linear_model import LogisticRegression
logi=LogisticRegression()
```

```python
logi.fit(x_train,y_train)
```

```python
y_pred=logi.predict(x_test)
```

### Calculating Classification_report and Confusion_matrix



```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print("Confusion_Matrix \n",confusion_matrix(y_test,y_pred))
```

## KNN = 66%

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
```



### Calculating Classification_report and Confusion_matrix



```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print("Confusion_Matrix \n",confusion_matrix(y_test,y_pred))
```

# ** Heart Disease Dataset Analysis Summary**



Exploratory Data Analysis (EDA) Insights:
1. The dataset spans a wide age range from 20 to 80 years, with a mean age of approximately 52.38 years and a standard deviation of 13.04 years. This indicates a diverse representation of age groups in the dataset.
2. Male individuals account for approximately 74.86% of the dataset, while female individuals represent about 25.14%. This suggests a higher proportion of male participants in the study.
3. Resting Blood Pressure (RestingBP): The mean resting blood pressure is approximately 138.70, with a standard deviation of 24.62. The range of blood pressure values indicates variability in individuals' cardiovascular health status.
4. Cholesterol Levels: The mean cholesterol level is approximately 253.52, with a standard deviation of 112.72. The wide range of cholesterol values underscores the diversity in lipid profiles among individuals.
5. Maximum Heart Rate Achieved (MaxHR): The mean maximum heart rate achieved during exercise is approximately 144.04, with a standard deviation of 28.45. This metric provides insights into individuals cardiovascular fitness and response to physical activity.

The dataset contains information on the presence or absence of heart disease. Approximately 54.84% of individuals have been diagnosed with heart disease, while 45.16% show no evidence of the condition.
